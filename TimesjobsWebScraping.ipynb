```python
from bs4 import BeautifulSoup
import requests
```

```python
url= 'https://www.timesjobs.com/candidate/job-search.html?searchType=personalizedSearch&from=submit&searchTextSrc=&searchTextText=&txtKeywords=python&txtLocation='
html_text= requests.get(url).text
soup = BeautifulSoup(html_text, 'lxml')
```

```python
#find the first company in the list offering job along with required skills and published date
job1= soup.find('li', class_='clearfix job-bx wht-shd-bx')
company1=job1.find('h3',class_='joblist-comp-name').text.replace(' ','')
skills1=job1.find('span',class_='srp-skills').text
published_date1= job1.find('span',class_= 'sim-posted').text

print(f"Company Name: {company1.strip()}")
print(f"Required Skills: {skills1.strip()}")
print(f"Job publish date: {published_date1}")
```

```python
#find all the companies in the list offering job along with required skills and published date
jobs= soup.find_all('li', class_='clearfix job-bx wht-shd-bx')
for job in jobs:
    company=job.find('h3',class_='joblist-comp-name').get_text(strip=True).replace(' ','')
    skills=job.find('span',class_='srp-skills').get_text(strip=True)
    published_date= job.find('span',class_= 'sim-posted').get_text(strip=True)
    
    print(f"Company Name: {company1.strip()}")
    print(f"Required Skills: {skills1.strip()}")
    print(f"Job publish date: {published_date1}")
```

```python
#find all the companies in the list offering job along with required skills and published date whivh have been posted a few days ago only
jobs= soup.find_all('li', class_='clearfix job-bx wht-shd-bx')
for job in jobs:
    published_date= job.find('span',class_= 'sim-posted').get_text(strip=True)
    if 'few' in published_date:
        company=job.find('h3',class_='joblist-comp-name').get_text(strip=True).replace(' ','')
        skills=job.find('span',class_='srp-skills').get_text(strip=True)
        
        print(f"Company Name: {company1.strip()}")
        print(f"Required Skills: {skills1.strip()}")
        print(f"Job publish date: {published_date1}")
        print('')
```

```python
#find all the companies in the list offering job along with their job description
jobs= soup.find_all('li', class_='clearfix job-bx wht-shd-bx')
for job in jobs:
    company=job.find('h3',class_='joblist-comp-name').get_text(strip=True).replace(' ','')
    url= job.header.h2.a['href']
    more_info_html= requests.get(url).text
    soup2 = BeautifulSoup(more_info_html, 'lxml')
    job_desc=soup2.find('code',class_='language-markdown').text
    print(f"Company Name: {company.strip()}")
    print(f"Job Description: {job_desc}")
    
```

```python
#find all the companies in the list offering job filtering by the skill entered by user
filter=input("Enter skill: ").strip()
print(f"Filtering by {filter}")
jobs= soup.find_all('li', class_='clearfix job-bx wht-shd-bx')
for job in jobs:
    company=job.find('h3',class_='joblist-comp-name').get_text(strip=True).replace(' ','')
    skills=job.find('span',class_='srp-skills').get_text(strip=True)
    if filter.lower() in skills.lower():        
        print(f"Company Name: {company1.strip()}")
        print(f"Required Skills: {skills1}")
        print('')
```
